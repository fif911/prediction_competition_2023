```python
import pandas as pd
import os

# Directory where the CSV files are stored
input_dir = 'data_dyad_monthly v0.6'

# List all CSV files in the directory
csv_files = [f for f in os.listdir(input_dir) if f.endswith('.csv')]
csv_files = list(sorted(csv_files))

# Verify that all months are present without missing months
# get min and max from cm_features
cm_features = pd.read_csv('data/cm_features_v0.6.csv')
min_month = cm_features['month_id'].min()
max_month = cm_features['month_id'].max()
del cm_features

print(f'Min month: {min_month}, Max month: {max_month}, diff: {max_month - min_month + 1}')
print(f'Number of CSV files: {len(csv_files)}')
```

```python
months = [f.split('_')[2][:3] for f in csv_files]
months = list(sorted(months))
expected_months = [str(i).zfill(2) for i in range(min_month, max_month + 1)]
assert months == expected_months, 'Missing months'
months == expected_months
```

```python

# Initialize an empty list to store DataFrames
dfs = []

# Loop over the list of CSV files
for file in sorted(csv_files):  # Sorting ensures the files are processed in order
    print(f'Reading {file}')
    file_path = os.path.join(input_dir, file)
    # Read the current CSV into a DataFrame
    df = pd.read_csv(file_path)
    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames in the list into one
df = pd.concat(dfs, ignore_index=True)
# save to parquet
df.to_parquet('data_dyad_monthly_v0.6.parquet', index=False)
print("Done!")
```
