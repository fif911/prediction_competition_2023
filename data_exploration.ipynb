```python
import pandas as pd

# data/cm_features_with_name.csv
cm_features = pd.read_csv('data/cm_features_with_name.csv')
cm_features.head(10)
```

```python
# filter Ukraine
cm_features_ukraine = cm_features[cm_features['country_id'] == 57]
cm_features_ukraine
```

```python
# plot correlation matrix for the whole dataset
import seaborn as sns
import matplotlib.pyplot as plt

# plot correlation matrix for the whole dataset for numerical features
drop = ['month_id', 'country_id', 'country', 'date']

cm_features_corr = cm_features.drop(drop, axis=1).corr()

```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(cm_features_corr, annot=False, cmap='coolwarm')
plt.show()
```
